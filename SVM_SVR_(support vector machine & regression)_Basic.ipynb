# TEST function by using Boston House Dataset

## Basic usage for predict

```python
from sklearn.datasets import load_boston
from sklearn.svm import SVR
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
boston = load_boston()
X = boston.data
Y = boston.target

# use rbf kernel
clf = SVR(kernel='rbf', C=1e3, gamma=0.1) # setting SVR model
clf.fit(X,Y)

# Predict the second data
predict_y=clf.predict(boston.data[[2]])
print(clf.score(X, Y))

plt.scatter(clf.predict(X),Y,s=2)
plt.plot(predict_y, predict_y, 'ro')
plt.plot([Y.min(), Y.max()], [Y.min(), Y.max()], 'k--', lw=2)
plt.xlabel('Predicted')
plt.ylabel('Measured')
```

```python
# Export the SVR machine
from sklearn.externals import joblib
joblib.dump(clf,"./machine_SVR.pkl")

# Import existing files
clf=joblib.load("./machine_SVR.pkl")
```

## 原理講解

支持向量機回歸，(SVR，support vector regression)，是支持向量機(SVM，support vector machine)對於回歸分析的應用。支持向量機是一種二分類模型，在特徵空間中以找到超平面的方式，將數值最大間隔化，藉由將數據映射至高維度空間，處理非線性分類問題。


主要使用的rbf為，徑向基核函數且最常被使用，能較其他其他核函數將數據提升至無窮維度的等級，使支援向量機的回歸結果在分析非線性數據時，得到較高的準確度。以下以多項式(Poly)核函數與徑向基核函數(RBF)比較：

從RBF核函數中的泰勒展開式其實為0至n維的多項式函數的和，此展開式帶入後使數據映射到無窮高的維度(特徵空間)，得到一個可與所有數據的殘差最小的帶狀平面，再對帶狀區域外的點位進行回歸。

```python
from sklearn.svm import SVC
model = SVC()
model.fit(x,y)
# test
model.predict([[-3,2]])
```

# Use Iris data to do SVM

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
iris = load_iris()
X = iris.data
Y = iris.target

x_train, x_test, y_train, y_test = train_test_split(X,Y,
                                                    test_size = 0.2,
                                                    random_state=11
                                                   )
```

```python
plt.scatter(x_train[:,0],x_train[:,1], c=y_train)
```

```python
from sklearn.svm import SVC
model = SVC()
model.fit(x_train,y_train)
y_test_predict = model.predict(x_test)
plt.scatter(x_test[:,0],x_test[:,1], c = model.predict(x_test) )
```

```python
# 看哪點預測錯誤
plt.scatter(x_test[:,0],x_test[:,1], c=model.predict(x_test) - y_test)
```
